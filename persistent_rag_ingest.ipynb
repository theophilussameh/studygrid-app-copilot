```python
from dotenv import load_dotenv
load_dotenv()

from ingest import load_faq_data, build_index
from rag_helper import RAGB

from openai import OpenAI
import os

from sqlitesearch import TextSearchIndex
```

```python


sqlite_index  = TextSearchIndex (
 text_fields=['question', 'question_ar', 'answer', 'answer_ar'],
 keyword_fields=['section'],
 db_path="data/faq.db"
)
```

```python
sqlite_index.count()
```

```python
documents = load_faq_data()
print(f"Loaded {len(documents)} documents")
```

```python

index = build_index(documents)
```

```python
openai_client = OpenAI(
    api_key=os.getenv("GROQ_API_KEY"),
    base_url="https://api.groq.com/openai/v1"
)
```

```python
assistant = RAGB( index=sqlite_index, openai_client=openai_client)

answer = assistant.rag("إزاي أنضم لجروب موجود؟")
print(answer)
```
